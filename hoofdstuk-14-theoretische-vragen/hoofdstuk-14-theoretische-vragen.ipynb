6. (Oefening 2 uit het boek) Beschouw een CNN bestaande uit drie convolu-tionele lagen. Elke laag gebruikt een3 × 3“kernel”, een “stride” gelijk aan2en “same” “padding”.

    De eerste laag produceert 100 “feature maps”, de middelste laag produ-ceert er 200, en de uitvoerlaag produceert er 400. De invoerbestanden zijn RGB-afbeeldingen met die 200 × 300 pixels groot zijn.

- Hoeveel parameters heeft deze CNN in totaal? Toon tussenstappen.


Formule per laag:
`(kernel_h × kernel_w × input_channels + 1) × output_channels`

```python
# Invoer
input_shape = (200, 300, 3)

# Laag 1
# Output shape: (200/2, 300/2, 100) = (100, 150, 100)
# Parameters (of weights): (3*3*3 + 1) * 100 = 2800
L1_params = (3*3*3 + 1) * 100  # 2800

# Laag 2
# Output shape: (100/2, 150/2, 200) = (50, 75, 200)
# Parameters (of weights): (3*3*100 + 1) * 200 = 180200
L2_params = (3*3*100 + 1) * 200  # 180200

# Laag 3
# Output shape: (50/2, 75/2, 400) = (25, 38, 400)
# Parameters (of weights): (3*3*200 + 1) * 400 = 720400
L3_params = (3*3*200 + 1) * 400  # 720400
print(f"L1 parameters: {L1_params}, L2 parameters: {L2_params}, L3 parameters: {L3_params}")

# Totaal aantal parameters
total_params = L1_params + L2_params + L3_params  # 903400
print(f"Totaal aantal parameters: {total_params}")

"""
L1 = (100,150,100) en L1 (gewicht) = (3 * 3 * 3 + 1) * 100 : 2800 ->
L2 = (50,75,200) en L2 (gewicht) = (3 * 3 * 100 + 1) * 200 : 180 200 ->
L3 = (25,38,400) en L3 (gewicht) = (3 * 3 * 200 + 1) * 400 : 720 400 ->
"""
```
